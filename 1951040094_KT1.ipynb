```python
def check_number(n):
  for i in range(2,n):
     if n % i == 0:
       return False
       return True
while True:
   n = int(input("Hãy nhập vào 1 số nguyên tố duy nhất")) 
   if check_number(n) == False:
     print("Xin hãy nhập lại:")
   else:
     break
```

```python
import math
def fibonacci(n):
    if (n < 0):
        return -1;
    elif (n == 0 or n == 1):
        return n;
    else:
        return fibonacci(n - 1) + fibonacci(n - 2);
n = int(input("Nhap vao so nguyen duong n = "));
print ("Tat ca cac so fibonacci nho hon",n);
i = 0;
fin = fibonacci(i);
while(fin < n):
    fin = fibonacci(i);
    if (fin < n):
        print(fin)
    i = i + 1;
```
